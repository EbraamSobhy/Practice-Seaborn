```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
```

```python
mart = pd.read_excel("supermarket_sales.xlsx")
mart.head()
```

```python
sns.violinplot(y='Total', data=mart)
```

```python
sns.violinplot(x='Total', data=mart)
```

```python
sns.violinplot(x='Payment', y='Total', hue='Gender', data=mart, split=True, inner='quartile')
```

```python
sns.violinplot(x='Payment', y='Total', hue='Gender', data=mart, split=True, inner='stick')
```

```python
sns.violinplot(x='Payment', y='Total', hue='Gender', data=mart, split=True, inner='quartile', bw=1.2)
```

```python
sns.violinplot(x='Payment', y='Total', hue='Gender', data=mart, split=True, inner='quartile', bw=1.2, cut=0)
```
