```python
l = [5, 12, 10, 14, 43, 11, 7, 9, 6]
```

```python
list1 = []

for item in l:
    if item % 2 == 0:
        list1.append(item)
    print("The final list using a for loop:", list1)
```

```python
l = [5, 12, 10, 14, 43, 11, 7, 9, 6]
list1 = [item for item in l if item % 2 == 0]
print("list:", list1)
```
